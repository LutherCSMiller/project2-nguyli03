# A story on online courses: Not a lot of people finish courses they register for!

## Abstract: 
Online courses have gained its popularity recent year. People can just go to websites like Coursera and edX to find a course that they one. In a CV, it is not uncommon to see someone listing out some online courses that he or she has done. However, I always wonder how many people actually finish an online courses. Online courses usually last for 6-8 weeks, which are about 2 months of commitment. Therefore, I decided to look into this problem.
